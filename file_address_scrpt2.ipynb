```python
import pandas as pd
from pathlib import Path

# 1. 读取 Excel
xlsx_path = "/Users/fantastic-lin/Documents/Andrew/aet21.github.io-master/software/software_info_list.xlsx"  # 换成你的文件名
df = pd.read_excel(xlsx_path)
df = df.fillna("")  # 避免 NaN

# 2. 输出目录（Jekyll collection 一般是 _software）
out_dir = Path("/Users/fantastic-lin/Documents/Andrew/aet21.github.io-master/_software/")
out_dir.mkdir(exist_ok=True)

def sanitize_filename(name: str) -> str:
    """简单处理文件名：去空格、换成下划线、全小写"""
    name = name.strip()
    if not name:
        name = "software"
    name = name.replace(" ", "_").lower()
    return name

def escape_quotes(s: str) -> str:
    """保证 front matter 里双引号不会破坏 YAML"""
    return s.replace('"', '\\"')

for _, row in df.iterrows():
    abbr   = str(row["abbreviation"]).strip()
    title  = str(row["wholeName"]).strip()
    state  = str(row["state"]).strip()
    link1  = str(row["link1"]).strip()
    link1Site = str(row["link1Site"]).strip()
    link2  = str(row["link2"]).strip()
    link2Site = str(row["link2Site"]).strip()
    imagePath = str(row["imagePath"]).strip()
    intro  = str(row["simpleIntroduction"]).strip()

    # 文件名用缩写（没有就用 wholeName）
    base_name = abbr if abbr else title
    filename = sanitize_filename(base_name) + ".md"
    filepath = out_dir / filename

    # 转义双引号
    abbr_q   = escape_quotes(abbr)
    title_q  = escape_quotes(title)
    state_q  = escape_quotes(state)
    link1_q  = escape_quotes(link1)
    link1Site_q = escape_quotes(link1Site)
    link2_q  = escape_quotes(link2)
    link2Site_q = escape_quotes(link2Site)
    image_q  = escape_quotes(imagePath)
    intro_q  = escape_quotes(intro)

    # 3. 生成 md front matter（按你给的模板）
    md_text = f"""---
abbreviation: "{abbr_q}"
categories:
  - software
wholeName: "{title_q}"
state: "{state_q}"
link1: "{link1_q}"
link1Site: "{link1Site_q}"
link2: "{link2_q}"
link2Site: "{link2Site_q}"
image: "/software/software_images/{image_q}"
simpleIntroduction: "{intro_q}"
---
"""

    filepath.write_text(md_text, encoding="utf-8")
    print(f"Wrote {filepath}")

print("✅ 所有 .md 已生成到 _software/ 目录。")
```

```python
import os
import re
from datetime import datetime

# 你的论文目录
PAPERS_DIR = "/Users/fantastic-lin/Documents/Andrew/aet21.github.io-master/_papers/"

# 月份英文转数字
MONTH_MAP = {
    "January": 1,
    "February": 2,
    "March": 3,
    "April": 4,
    "May": 5,
    "June": 6,
    "July": 7,
    "August": 8,
    "September": 9,
    "October": 10,
    "November": 11,
    "December": 12
}

def process_markdown_file(file_path):
    with open(file_path, "r", encoding="utf-8") as f:
        lines = f.readlines()

    year = None
    month = None
    front_matter_end = None
    has_date = False

    # 读取 front matter
    for i, line in enumerate(lines):
        line = line.strip()
        if line == "---":
            if front_matter_end is None:
                front_matter_end = 0
            else:
                front_matter_end = i
                break
        if line.startswith("year:"):
            year = line.split(":", 1)[1].strip()
        if line.startswith("month:"):
            month_str = line.split(":", 1)[1].strip()
            month = MONTH_MAP.get(month_str)
        if line.startswith("date:"):
            has_date = True

    if not year or not month:
        print(f"⚠️  跳过 {os.path.basename(file_path)}（缺少 year 或 month）")
        return

    if has_date:
        print(f"✅ 已存在 date，跳过 {os.path.basename(file_path)}")
        return

    # 生成 date
    date_str = f"date: {year}-{month:02d}-01\n"

    # 插入到 front matter 里（在 month 后面）
    new_lines = []
    inserted = False
    for line in lines[:front_matter_end]:
        new_lines.append(line)
        if not inserted and line.strip().startswith("month:"):
            new_lines.append(date_str)
            inserted = True
    new_lines.extend(lines[front_matter_end:])

    # 写回文件
    with open(file_path, "w", encoding="utf-8") as f:
        f.writelines(new_lines)

    print(f"✅ 成功写入 {os.path.basename(file_path)} | {date_str.strip()}")

# 遍历所有 md 文件
for filename in os.listdir(PAPERS_DIR):
    if filename.endswith(".md"):
        process_markdown_file(os.path.join(PAPERS_DIR, filename))

print("\n🎉 所有文件处理完成！")
```
