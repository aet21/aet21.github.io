```python
import pandas as pd
from pathlib import Path

# 1. 读取 Excel
xlsx_path = "/Users/fantastic-lin/Documents/Andrew/aet21.github.io-master/software/software_info_list.xlsx"  # 换成你的文件名
df = pd.read_excel(xlsx_path)
df = df.fillna("")  # 避免 NaN

# 2. 输出目录（Jekyll collection 一般是 _software）
out_dir = Path("/Users/fantastic-lin/Documents/Andrew/aet21.github.io-master/_software/")
out_dir.mkdir(exist_ok=True)

def sanitize_filename(name: str) -> str:
    """简单处理文件名：去空格、换成下划线、全小写"""
    name = name.strip()
    if not name:
        name = "software"
    name = name.replace(" ", "_").lower()
    return name

def escape_quotes(s: str) -> str:
    """保证 front matter 里双引号不会破坏 YAML"""
    return s.replace('"', '\\"')

for _, row in df.iterrows():
    abbr   = str(row["abbreviation"]).strip()
    title  = str(row["wholeName"]).strip()
    state  = str(row["state"]).strip()
    link1  = str(row["link1"]).strip()
    link1Site = str(row["link1Site"]).strip()
    link2  = str(row["link2"]).strip()
    link2Site = str(row["link2Site"]).strip()
    imagePath = str(row["imagePath"]).strip()
    intro  = str(row["simpleIntroduction"]).strip()

    # 文件名用缩写（没有就用 wholeName）
    base_name = abbr if abbr else title
    filename = sanitize_filename(base_name) + ".md"
    filepath = out_dir / filename

    # 转义双引号
    abbr_q   = escape_quotes(abbr)
    title_q  = escape_quotes(title)
    state_q  = escape_quotes(state)
    link1_q  = escape_quotes(link1)
    link1Site_q = escape_quotes(link1Site)
    link2_q  = escape_quotes(link2)
    link2Site_q = escape_quotes(link2Site)
    image_q  = escape_quotes(imagePath)
    intro_q  = escape_quotes(intro)

    # 3. 生成 md front matter（按你给的模板）
    md_text = f"""---
abbreviation: "{abbr_q}"
categories:
  - software
wholeName: "{title_q}"
state: "{state_q}"
link1: "{link1_q}"
link1Site: "{link1Site_q}"
link2: "{link2_q}"
link2Site: "{link2Site_q}"
image: "/software/software_images/{image_q}"
simpleIntroduction: "{intro_q}"
---
"""

    filepath.write_text(md_text, encoding="utf-8")
    print(f"Wrote {filepath}")

print("✅ 所有 .md 已生成到 _software/ 目录。")
```
